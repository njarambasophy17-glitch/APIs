```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'https://www.scrapethissite.com/pages/simple/'
```

```python
response = requests.get(url)
print(response.text)
```

```python
def extract_article_title(url):
  response = requests.get(url)
  soup = BeautifulSoup(response.text, 'html.parser')
  title_element = soup.find('h1')

  if title_element:
    title = title_element.get_text(strip=True)
  else:
    print('Title not found')
    title = None # Assign None when title is not found
  return title
```

```python
def extract_country_info(url):
  response = requests.get(url)
  soup = BeautifulSoup(response.text, 'html.parser')
  country_elements = soup.find_all('div', class_ = 'col-md-4 country')
  country_data = []
  for country_element in country_elements:
    name_element = country_element.find('h3', class_ = 'country-name')
    info_element = country_element.find('div', class_ = 'country-info')

    name = name_element.get_text(strip=True) if name_element else None
    info = info_element.get_text(strip=True) if info_element else None

    country_data.append({'name': name, 'info': info})
  return country_data
```

```python
def link_to_article(url):
  response = requests.get(url)
  soup = BeautifulSoup(response.text, 'html.parser')
  link_to_article = soup.find('a', class_ ='a href' )
  if link_to_article:
    link = link_to_article.get('href')
    return link
```

```python
def __main__():
  article_title = extract_article_title(url)
  country_data = extract_country_info(url)


  print("Article Title:", article_title)
  # print("Heading:", heading)
  # print("Article Text:", article_text)
  # print("Download Link:", link)
  print("Country Data:", country_data)

__main__()
```
